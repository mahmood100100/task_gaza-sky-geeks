```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv("/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv")
print(movies.head())
```

```python
#if we apply the code line under then we cant use it again because the columns will removed
movies.drop(['Poster_Link','Star2','Star3','Star4','Gross'],axis=1,inplace=True)
movies.dropna()
movies.fillna('UnKnown')
print(movies)
```

```python
the_most_popular_genre=movies.Genre.value_counts().idxmax()
print('the most popular genre is :',the_most_popular_genre,"\n__________________________")
```

```python
count_movies_for_genre=movies.Genre.value_counts()
print('count movies for genre ')
print(count_movies_for_genre,"\n___________________________")
```

```python
plt.figure(figsize=(20, 10))
count_movies_for_genre.plot.bar()
plt.xlabel('Genre')
plt.ylabel('pobular_rate')
plt.title('genres with it movies number')
plt.show()
```

```python
count_movies_for_director=movies.Director.value_counts().reset_index()
count_movies_for_director.columns = ['Director', 'mCount']
print(count_movies_for_director)
```

```python
count_movies_for_director.plot.bar()
plt.xlabel('Director')
plt.ylabel('top_rated')
plt.title('Director with the Most Top-rated Movies')
plt.show()
```
